## Unsupervised Machine Learning
This type of learning the model is trained on data that does not have labelled responses.The main is to uncover hidden patterns, structures or relationships within the data.

## Types of Unsupervised Learning
1. Clustering
The GOAL is to Group similar Data Points together based on their similar features. E,g Safaricom Customers have been grouped into:

    * SMS Services: Send Promotional SMS updates that only target these customers.

    * Call ServicesUusupervisedMLClustering

    * Bundles Subscribers

Clustering using the Following Algorithms:
* K-Means: Where the groupings is done based on similarities of the Mean e,g Mean Age. This ONLY work on Numerical Data.

* K-Mode: Grouped based on Categorical data e,g Gender, Location...

* K-Prototype: Grouping is done based on BOTH Numerical(Mean) and Categorical(Mode)

Applications of Clustering Include:
1. Marget/Customer Segmentation.
2. Image Segmentation
3. Social Network Analyis

```python
# Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CUSTOMER SEGMENTATION ON MALL CUSTOMERS DATA
# KMEANS
# Reading the Data

customers_data = pd.read_csv('https://msi.martial.co.ke/data/Mall_Customers.csv')
customers_data.head()
```

```python
# EXPLANATORY DATA ANALYSIS
# a) Statistical Analysis
customers_data.describe()
```

```python
# Checking for Empties
customers_data.isnull().sum()
```

```python
# Drop the Empties
customers_data.dropna(inplace=True)

# Check Empties Again
customers_data.isnull().sum()
```

```python
# MACHINE LEARNING
# The Mall wanted to Cluster(Segment) their customers based on:
# a) Age: Because purchasing patterns depends on the age(behaviour)
# b) Annual Income: the purchasing power depends on the value of the customer(Value Based Segment)
# c) Spending Score: Depends on the Needs of the Customer(Need Based)
```

```python
# a) Get a subset(portion) that contains the columns of interest
customers_subset = customers_data[['Age', 'Annual Income', 'Spending Score']]
customers_subset.head()
```

```python
# b) Store the Columns as Features(X) in the Machine Learning Model as array
# NB: There is no Response Y(Labels)
array = customers_subset.values
X = array[:, :]

print(X.shape)
```

```python
# c) Import Scientific Algorithms(KMeans) that would clustr the data into Segments
# The Clustering is done by Grouping the Data Based on its Similar Data Points

from sklearn.cluster import KMeans
model = KMeans(n_clusters=5, random_state=42)
model.fit(X)
```

```python
# d) Gather the Centronoids
# Centroids: They refers representative points in dataset that acts as the cluster center
# E,g In terms of Age we can have 18(Youth), 35(Semi-Adults), 45(Adults

centronoids = model.cluster_centers_
centronoids_df = pd.DataFrame(centronoids, columns=['Age', 'Annual Income', 'Spending Score'])
centronoids_df
```

```python
# e) Generating Members of a Cluster
# Add Labels to the Original Dataset
customers_data['Cluster'] = model.labels_
customers_data.head()
```

```python
# Cluster 0 Members(Age-44.054054	Income-55.175676	Spending-49.297297)
cluster0 = customers_data[customers_data['Cluster']== 0]
cluster0
```

```python
# Age-32.783784	Income-86.378378	Spending-82.297297
cluster1 = customers_data[customers_data['Cluster']== 1]
cluster1.to_csv('cluster1.csv')
```
